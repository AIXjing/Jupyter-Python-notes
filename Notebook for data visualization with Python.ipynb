**There are three ways to plot data:**

1. `matplotlib.pyplot.plot()`: a library that Pandas use in their plot function

2. `pandas.DataFrame.plot()`: a convenient shortcut

3. `seaborn()`

## 1 Create a figure using matplotlib.pyplot

### 1.1 Create a figure by pyplot.figure()

`matplotlib.pyplot.figure()` creates a new figure, or activate an existing figur. [Check this link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html#matplotlib.pyplot.figure)

Return Figure Object

```python
# A simple example

%matplotlib inline
import matplotlib.pyplot as plt

# Create a figure
fig1 = plt.figure()
# Plot data in the current figure
plt.plot(range(5))
```

### 1.2 Create a figure by pyplot.axes()

`matplotlib.pyplot.axes()` adds an axes to the current figure and make it the current axes. [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axes.html#examples-using-matplotlib-pyplot-axes)

Returns: axes or a subclass of axes

```python
# Create a new full window axes
plt.axes()

# Or create a new axes with specified dimensions and some kwargs
# 4-tuple arguement corresponds to the position of the axes (left, bottom,width, height)

plt.axes((0.5, 0.5, 0.5, 0.5), facecolor='b')
```

### 1.3.3.3 A good example of how to use figure and axes properly

```python
np.random.seed(19680801)  # Fixing random state for reproducibility.

# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 10.0, dt)
r = np.exp(-t[:1000] / 0.05)  # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)] * dt  # colored noise

fig, main_ax = plt.subplots()
main_ax.plot(t, s)
main_ax.set_xlim(0, 1)
main_ax.set_ylim(1.1 * np.min(s), 2 * np.max(s))
main_ax.set_xlabel('time (s)')
main_ax.set_ylabel('current (nA)')
main_ax.set_title('Gaussian colored noise')

# this is an inset axes over the main axes
right_inset_ax = fig.add_axes([.65, .6, .2, .2], facecolor='b')
right_inset_ax.hist(s, 400, density=True)
right_inset_ax.set(title='Probability', xticks=[], yticks=[])

# this is another inset axes over the main axes
left_inset_ax = fig.add_axes([.2, .6, .2, .2], facecolor='k')
left_inset_ax.plot(t[:len(r)], r)
left_inset_ax.set(title='Impulse response', xlim=(0, .2), xticks=[], yticks=[])

plt.show()
```

## 2 Plot by pandas.Dataframe.plot()

[Detailed description](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html)

```python
# a simple example

import pandas as pd

df = pd.DataFrame({
    'X': range(5),
    'A': [1,10,20,30,40]
})

df['B'] = df['A']*2 + 0.1
df['C'] = df['B']%3

df.plot(x="X", y=["A", "B", "C"], kind="line")
```
